```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser 
from selenium import webdriver
import json
```

```python
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='content_title')

titles = []
for result in results:
    title = result.a.text
#     print(title)
    titles.append(title)
    
latest_title = titles[0]
print(latest_title)
```

```python
results = soup.find_all('div', class_="rollover_description_inner")

news_p = []
for result in results:
    p = result.text
#     print(p)
    news_p.append(p)
    
latest_news_p = news_p[0]

print(latest_news_p)  
```

```python
url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
```

## Mars Hemispheres

```python
astropedia_url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
```

```python
browser= Browser('chrome')
browser.visit(astropedia_url)
```

```python
links = browser.find_by_css("a.product-item h3")
hemisphere_urls = []
for i in range(len(links)):
    hemisphere = {}
    browser.find_by_css("a.product-item h3")[i].click()
    sample = browser.find_link_by_text('Sample').first
    hemisphere['title'] = browser.find_by_css('h2.title').text
    hemisphere['image_url'] = sample['href']
    
#     title = browser.find_by_css('h2.title').text
#     hemisphere[title] = original['href']
    hemisphere_urls.append(hemisphere)

    browser.back()

hemisphere_urls


```

## Scrape Function

```python
def scrape():
    information = {'title': latest_title,
            'summary': latest_news_p,
            'image': featured_image_url,
            'post': latest_post,
            'facts': facts_table,
            'hemispheres': hemisphere_urls
    }
    return information
```
